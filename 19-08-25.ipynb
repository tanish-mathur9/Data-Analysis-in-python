```python
my_rng = np.random.default_rng(seed= 1857) #to generate random no.
print(my_rng) #bitwise generator
time_spent_AMZN = my_rng.integers(100, 1000, size= 100)
# time_spent_AMZN = my_rng.integers(100, 1000, size= int(1e6))
print (f'Time spent on AMZN in seconds by {len(time_spent_AMZN)} shoppers: \n {time_spent_AMZN}')
# print (time_spent_AMZN)
```
