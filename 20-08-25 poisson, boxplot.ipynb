```python
np.random.seed(1857)
pois_20m_100n = np.random.poisson(20,100)
_,_, hist_fig = plt.hist (pois_20m_100n, bins = len(pois_20m_100n), color= "b") # only histogram is printed

plt.axvline(pois_20m_100n.mean(), color= "r", ls ="-.") #for vertical line
p20_mean = pois_20m_100n.mean()
# plt.text(x= 20*1.01, y= 10, s= f"Mean is: {pois_20m_100n.mean()}" ) #hardcoding should not
# plt.text(x= p20_mean*0.65, y= 10, s= f"Mean is: {p20_mean}" ) #hardcoding should not
plt.text(x= p20_mean*0.65, y= 10, s= f"Mean is: {p20_mean}", color= "r" ) #to print left side 
```

## np.quantile() for percentile

```python
bumrah_ts_views = 60- pois_20m_100n #will get the left skewed distribution
# np.set_printoptions(legacy = '1.25')

Q1 = np.quantile (bumrah_ts_views, 0.25)
Q2 = np.quantile (bumrah_ts_views, 0.5)
Q3 = np.quantile (bumrah_ts_views, 0.75)

Q1, Q2, Q3
```
