Q1->  generate an array of no.s from 20-45 in steps of 0.5

```python
# op = list(range(20, 45, 0.5)) #'float' object cannot be interpreted as an integer
```

```python
import numpy as np
```

```python
# np.linspace(20, 45, 0.5) #error
op = np.arange(20, 45, 0.1) 
len(op) #250
shape_ = op.shape #(250,) ...........for general purpose
op.shape[0] #250 (scalar) ...........for general purpose
```

## a in arange is array 

```python
import matplotlib.pyplot as plt #importing pyplot module from matplotlib library as plt
```

```python
plt.hist(op, bins = 250)
# plt.show()
```

```python
np.random.seed(2000000) #for reproducibility
np.random.uniform(low=20, high=45, size=200) #named arguments
rand_utif_2c = np.random.uniform(20, 45, 200) #positional arguments
# np.random.uniform(lo=1, hi=100, size=100) #throws an error
rand_utif_2c
```

```python
plt.hist(rand_utif_2c, bins = 200, color = "red") #har value ka apna ek bar bnta h using bins
# bins are the class intervals
plt.show
```

```python
np.random.seed(1857)
rand_gauss_2c = np.random.normal(loc= 100, scale = 10, size = int(1e6))
rand_gauss_2c.sort()
plt.hist(rand_gauss_2c, edgecolor = "white", color= "black")
plt.title("")
```

## loc -> location of mean
## scale -> standard deviation
## size -> 2000
